```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
configurations=["1","2","3","4","6","7","1A","2A","3A","4A","6A","7A"]
```

```python
rdata = pd.DataFrame(columns=['configurations', 'errors'])
for i in range(0,len(configurations)):
    file = r'C:\Users\GagaN\Desktop\Analysis\testingADW\CPU-'
    file=file+configurations[i]+r'\500.csv'
    data= pd.read_csv (file)
    df = pd.DataFrame(data, columns= ['responseCode','success','Latency','Connect'])
    rdata=rdata.append({'configurations': configurations[i] , 'errors': df.success.value_counts()[0]}, ignore_index=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = "configurations", y = "errors", palette = 'Reds',hue='errors', data = rdata,dodge=False).set_title("Errors for different configurations for 500 concurrent users")
#plt.show()
plt.savefig('errors.png')
```

```python
rdata = pd.DataFrame(columns=['configurations', 'errors'])
for i in range(0,len(configurations)):
    file = r'C:\Users\GagaN\Desktop\Analysis\testingADW\CPU-'
    file=file+configurations[i]+r'\500.csv'
    data= pd.read_csv (file)
    df = pd.DataFrame(data, columns= ['responseCode','success','Latency','Connect'])
    df=df[df['success'] != 0 ]
    df['Latency']=df['Latency']-df['Connect']
    df_mean = df['Latency'].mean()
    rdata=rdata.append({'configurations': configurations[i] , 'Latency': df_mean}, ignore_index=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = "configurations", y = "Latency", palette = 'Greens_r',hue='Latency', data = rdata,dodge=False).set_title("Latency for different configurations for 500 concurrent users")
#plt.show()
plt.savefig('latency500.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
configurations=["1","2","3","4","6","7","1A","2A","3A","4A","6A","7A","100t","200t"]
```

```python
rdata = pd.DataFrame(columns=['configurations', 'errors'])
for i in range(0,len(configurations)):
    file = r'C:\Users\GagaN\Desktop\ADW-Analysis\testingADW\CPU-'
    file=file+configurations[i]+r'\500.csv'
    data= pd.read_csv (file)
    df = pd.DataFrame(data, columns= ['responseCode','success','Latency','Connect'])
   # print(df.success.value_counts().tolist()[0])
    errors=500-df.success.value_counts().tolist()[0]
   # print(errors)
    rdata=rdata.append({'configurations': configurations[i] , 'errors': errors}, ignore_index=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = "configurations", y = "errors", palette = 'Reds',hue='errors', data = rdata,dodge=False).set_title("Service Unavailable errors for different configurations for 500 concurrent users")
#plt.show()
plt.savefig('errors.png')
```

```python
rdata = pd.DataFrame(columns=['configurations', 'errors'])
for i in range(0,len(configurations)):
    file = r'C:\Users\GagaN\Desktop\ADW-Analysis\testingADW\CPU-'
    file=file+configurations[i]+r'\1000.csv'
    data= pd.read_csv (file)
    df = pd.DataFrame(data, columns= ['responseCode','success','Latency','Connect'])
    errors=1000-df.success.value_counts().tolist()[0]
    rdata=rdata.append({'configurations': configurations[i] , 'errors':errors}, ignore_index=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = "configurations", y = "errors", palette = 'Reds',hue='errors', data = rdata,dodge=False).set_title("Service Unavailable errors for different configurations for 1000 concurrent users")
#plt.show()
plt.savefig('errors1000.png')
```

```python
rdata = pd.DataFrame(columns=['configurations', 'errors'])
for i in range(0,len(configurations)):
    file = r'C:\Users\GagaN\Desktop\ADW-Analysis\testingADW\CPU-'
    file=file+configurations[i]+r'\500.csv'
    data= pd.read_csv (file)
    df = pd.DataFrame(data, columns= ['responseCode','success','Latency','Connect'])
    df=df[df['success'] != 0 ]
    df['Latency']=df['Latency']-df['Connect']
    df_mean = df['Latency'].mean()
    rdata=rdata.append({'configurations': configurations[i] , 'Latency': df_mean}, ignore_index=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = "configurations", y = "Latency", palette = 'Greens_r',hue='Latency', data = rdata,dodge=False).set_title("Latency for different configurations for 500 concurrent users")
#plt.show()
plt.savefig('latency500.png')
```
